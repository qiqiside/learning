```python
from common import Sort
```

```python
class BubbleSort(Sort):
    
    def __init__(self, A = []):
        super(BubbleSort, self).__init__(A)
        
    def _sort(self):
        A = self.A
        N = len(A)
        
        for i in range(N - 1):
            flag = True
            
            for j in range(N - 1 - i):
                if A[j] > A[j + 1]:
                    A[j], A[j + 1] = A[j + 1], A[j]
                    flag = False
            
            if flag:
                break
            
        return self.A
```

```python
BubbleSort().verify()
```
