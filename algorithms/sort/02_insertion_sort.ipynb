```python
from common import Sort
```

```python
class InsertionSort(Sort):
    
    def __init__(self, A = []):
        super(InsertionSort, self).__init__(A)
    
    def insert(self, pos, value):
        A = self.A
        i = pos - 1
        
        while i >= 0 and A[self.start + i] > value:
            A[self.start + i + 1] = A[self.start + i]
            i -= 1
        
        A[self.start + i + 1] = value
    
    def _sort(self):
        A = self.A
        
        for i in range(1, self.end - self.start + 1):
            self.insert(i, A[self.start + i])
        
        return self.A
```

```python
InsertionSort().verify()
```
