```python
from common import Sort
```

```python
class ShellSort(Sort):
    
    def __init__(self, A = []):
        super(ShellSort, self).__init__(A)
    
    def _sort(self):
        A = self.A
        N = self.end - self.start + 1
        
        h = 1
        while h < N // 3:
            h = 3 * h + 1
        
        while h >= 1:
            for i in range(h, N):
                for j in range(self.start + i, self.start + h - 1, -h):
                    if A[j] < A[j - h]:
                        A[j], A[j - h] = A[j - h], A[j]
                    else:
                        break
                
            h = h // 3

        return self.A
```
