```python
class InsertionSort(Sort):
    
    def __init__(self, A = []):
        super(InsertionSort, self).__init__(A)
    
    def insert(self, pos, value):
        A = self.A
        i = pos - 1
        
        while i >= 0 and A[self.start + i] > value:
            A[self.start + i + 1] = A[self.start + i]
            i -= 1
        
        A[self.start + i + 1] = value
    
    def _sort(self):
        A = self.A
        
        for i in range(1, self.end - self.start + 1):
            self.insert(i, A[self.start + i])
        
        return self.A
    
    
class QuickSortWithInsertion(Sort):
    
    def __init__(self, A = []):
        super(QuickSortWithInsertion, self).__init__(A)
        self.M = 5
    
    def quicksort(self, p, r):
        if r <= p + self.M:      
            InsertionSort().sort(self.A, p, r)
            return
        
        q = self.partition(p, r)
        self.quicksort(p, q - 1)
        self.quicksort(q + 1, r)
        
    def partition(self, p, r):
        A = self.A
        v = A[p]
        
        i = p + 1
        j = r
        while True:
            while A[i] < v:
                if i == r:
                    break
                i += 1
            
            while v < A[j]:
                if j == p:
                    break
                j -= 1
        
            if i >= j:
                break
                
            A[i], A[j] = A[j], A[i]
            
        A[p], A[j] = A[j], A[p]
        return j
    
    def _sort(self):
        self.quicksort(self.start, self.end)        
        return self.A
    
```

```python
QuickSortWithInsertion().verify()
```
