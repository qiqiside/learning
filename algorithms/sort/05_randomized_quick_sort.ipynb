```python
from common import Sort
import random
```

```python
class RandomizedQuickSort(Sort):
    
    def __init__(self, A = []):
        super(RandomizedQuickSort, self).__init__(A)
    
    def randomizedQuickSort(self, p, r):
        if p < r and p >= 0 and r < self.end - self.start + 1:
            q = self.randomizedPartition(p, r)
            self.randomizedQuickSort(p, q - 1)
            self.randomizedQuickSort(q + 1, r)
    
    def randomizedPartition(self, p, r):
        i = random.randint(p, r)
        A = self.A
        
        A[r], A[i] = A[i], A[r]
        
        return self.partition(p, r)
    
    def partition(self, p, r):
        A = self.A
        x = A[r]
        i = p - 1
        
        for j in range(p, r):
            if A[j] <= x:
                i += 1
                
                A[i], A[j] = A[j], A[i]
        
        A[i + 1], A[r] = A[r], A[i + 1]
        
        return i + 1

    def _sort(self):
        self.randomizedQuickSort(self.start, self.end)
        return self.A
        
```
