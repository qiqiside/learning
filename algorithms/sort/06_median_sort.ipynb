```python
from common import Sort
```

```python
class MedianSort(Sort):
    
    def __init__(self, A = []):
        super(MedianSort, self).__init__(A)
    
    def partition(self, lo, hi, pivotIndex):
        A = self.A
        pivot = A[pivotIndex]
        
        A[hi], A[pivotIndex] = A[pivotIndex], A[hi]
        
        store = lo
        for idx in range(lo, hi):
            if A[idx] < pivot:
                A[idx], A[store] = A[store], A[idx]
                store += 1
        
        A[hi], A[store] = A[store], A[hi]
        
        return store
        
    def selectPivotIndex(self, lo, hi):
        return lo
    
    def selectKth(self, k, lo, hi):
        idx = self.selectPivotIndex(lo, hi)
        #print('k = %d, lo = %d, hi = %d' % (k, lo, hi))
        
        pivotIndex = self.partition(lo, hi, idx)
        
        #print('pivotIndex = %d' % (pivotIndex))
        
        if lo + k - 1 == pivotIndex:
            return pivotIndex
        elif lo + k - 1 < pivotIndex:
            return self.selectKth(k, lo, pivotIndex - 1)
        else:
            return self.selectKth(k - (pivotIndex - lo + 1), pivotIndex + 1, hi)
    
    def medianSort(self, lo, hi):
        if hi <= lo:
            return
        
        mid = (hi - lo) // 2
        me = self.selectKth(mid + 1, lo, hi)
        self.medianSort(lo, lo + mid - 1)
        self.medianSort(lo + mid + 1, hi)
        
    def _sort(self):
        self.medianSort(self.start, self.end)        
        return self.A
    
```
