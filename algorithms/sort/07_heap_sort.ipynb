```python
class HeapSort(Sort):
    
    def __init__(self, A = []):
        super(HeapSort, self).__init__(A)
    
    def parent(self, i):
        return self.start + i // 2
    
    def left(self, i):
        return self.start + 2 * i + 1
    
    def right(self, i):
        return self.start + 2 * i + 2

    def maxHeapify(self, i):
        if i >= self.heapSize:
            return
        
        l = self.left(i)
        r = self.right(i)
        
        if l < self.heapSize and self.A[l] > self.A[i]:
            largest = l
        else:
            largest = i
        
        if r < self.heapSize and self.A[r] > self.A[largest]:
            largest = r
        
        if largest != i:
            self.A[i], self.A[largest] = self.A[largest], self.A[i]
            self.maxHeapify(largest)
         
    def buildMaxHeap(self):
        self.heapSize = self.end - self.start + 1
        for i in range(self.heapSize // 2 - 1, -1, -1):
            self.maxHeapify(i)     

    def _sort(self):
        self.heapSize = None
               
        self.buildMaxHeap()

        for i in range(self.end - self.start, 0, -1):
            self.A[self.start], self.A[self.start + i] = self.A[self.start + i], self.A[self.start]
            self.heapSize -= 1
            self.maxHeapify(0)
        
        return self.A
    
```
