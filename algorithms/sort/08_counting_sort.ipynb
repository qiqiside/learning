```python
from common import Sort, gen_data
```

```python
class CountingSort(Sort):
    
    MAX_VALUE = 3
    
    def __init__(self, A = []):
        super(CountingSort, self).__init__(A)
    
    def gen_data(self, x):
        return gen_data(x, 0, self.MAX_VALUE)
    
    def _sort(self):
        A = self.A
        k = self.MAX_VALUE + 1
        B = [ 0 ] * k
        N = len(A)
        
        for i in range(N):
            B[A[i]] += 1
        
        idx = 0
        for i in range(k):
            while B[i] > 0:
                A[idx] = i
                idx += 1
                B[i] -= 1
        
        return self.A
```

```python
CountingSort().verify(10, 10)
```
