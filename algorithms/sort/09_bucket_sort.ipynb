```python
class InsertionSort(Sort):
    
    def __init__(self, A = []):
        super(InsertionSort, self).__init__(A)
    
    def insert(self, pos, value):
        A = self.A
        i = pos - 1
        
        while i >= 0 and A[self.start + i] > value:
            A[self.start + i + 1] = A[self.start + i]
            i -= 1
        
        A[self.start + i + 1] = value
    
    def _sort(self):
        A = self.A
        
        for i in range(1, self.end - self.start + 1):
            self.insert(i, A[self.start + i])
        
        return self.A
```

```python
class BucketSort(Sort):
    
    MAX_VALUE = 10000
    
    def __init__(self, A = []):
        super(BucketSort, self).__init__(A)
    
    def bucketHash(self, x):
        return x * len(self.A) // (self.MAX_VALUE + 1)
    
    def gen_data(self, num):
        return gen_data(num, 0, self.MAX_VALUE)
    
    def _sort(self):
        A = self.A
        N = len(A)
        
        # create buckets
        B = []
        for i in range(N):
            B.append([])
        
        # put elements to buckets
        for i in range(N):
            k = self.bucketHash(A[i])
            B[k].append(A[i])

        # extract elements from buckets
        idx = 0
        for i in range(N):
            InsertionSort(B[i]).sort()
            
            for m in range(0, len(B[i])):
                A[idx] = B[i][m]
                idx += 1
        
        return self.A
```
