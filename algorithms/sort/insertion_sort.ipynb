```python
class InsertionSort:
    
    def __init__(self, A):
        self.A = A
    
    def insert(self, pos, value):
        A = self.A
        i = pos - 1
        
        while i >= 0 and A[i] > value:
            A[i + 1] = A[i]
            i -= 1
        
        A[i + 1] = value
    
    def sort(self):
        A = self.A
        
        for i in range(1, len(A)):
            self.insert(i, A[i])
                
```
