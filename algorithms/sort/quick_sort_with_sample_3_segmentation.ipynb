```python
class QuickSortWithThreeSampleSegmentation(Sort):
    
    def __init__(self, A = []):
        super(QuickSortWithThreeSampleSegmentation, self).__init__(A)
    
    def quicksort(self, p, r):
        if r <= p:
            return
        
        q = self.partition(p, r)
        self.quicksort(p, q - 1)
        self.quicksort(q + 1, r)
        
    def valueSample3(self, p, r):
        A = self.A
        m = (p + r) // 2
        
        minimum = maximum = p
        
        if A[p] < A[m]:
            maximum = m
        else:
            mininum = m
        
        if A[maximum] <= A[r]:
            return maximum 
        elif A[minimum] >= A[r]:
            return minimum
        else:
            return r
        
    def partition(self, p, r):
        A = self.A
        middle = self.valueSample3(p, r)
                
        A[p], A[middle] = A[middle], A[p]
        
        v = A[p]
        i = p + 1
        j = r
        while True:
            while A[i] < v:
                if i == r:
                    break
                i += 1
            
            while v < A[j]:
                if j == p:
                    break
                j -= 1
        
            if i >= j:
                break
                
            A[i], A[j] = A[j], A[i]
            
        A[p], A[j] = A[j], A[p]
        return j
    
    def _sort(self):
        self.quicksort(self.start, self.end)        
        return self.A
    
```
