```python
from common import Sort
```

```python
class QuickSortWithThreeWaySplit(Sort):
    
    def __init__(self, A = []):
        super(QuickSortWithThreeWaySplit, self).__init__(A)
    
    def quicksort(self, lo, hi):
        if hi <= lo:
            return
        
        A = self.A
                
        lt = lo
        i = lo + 1
        gt = hi
        
        v = A[lo]
        
        while i <= gt:
            if A[i] < v:
                A[i], A[lt] = A[lt], A[i]
                i += 1
                lt += 1
            elif A[i] > v:
                A[i], A[gt] = A[gt], A[i]
                gt -= 1
            else:
                i += 1
        
        self.quicksort(lo, lt - 1)
        self.quicksort(gt + 1, hi)
    
    def _sort(self):
        self.quicksort(self.start, self.end)        
        return self.A
    
```

```python
QuickSortWithThreeWaySplit().verify()
```
