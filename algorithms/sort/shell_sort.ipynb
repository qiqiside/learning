```python
class ShellSort:
    
    def __init__(self, A):
        self.A = A
    
    def sort(self):
        A = self.A
        N = len(A)
        
        h = 1
        while h < N // 3:
            h = 3 * h + 1
        
        while h >= 1:
            for i in range(h, N):
                for j in range(i, h - 1, -h):
                    if A[j] < A[j - h]:
                        A[j], A[j - h] = A[j - h], A[j]
                
            h = h // 3
```

```python
shellSort = ShellSort(data)
shellSort.sort()

draw_array('sorted data:', data)
```
